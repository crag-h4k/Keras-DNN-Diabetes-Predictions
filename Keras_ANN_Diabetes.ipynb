```python
#https://machinelearningmastery.com/tutorial-first-neural-network-python-keras/
#Exercise of changing the complexity and sturcture of a ANN to increase accuracy of model
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
np.random.seed(7)
```

```python
#setting up df and input/output
dataset = np.loadtxt('pima-indians-diabetes.csv', delimiter=',')
X = dataset[:,0:8]
Y = dataset[:,8]

#creating model here
model = Sequential()
#Dense = fully connected layers; input_dim = sets amount of input vars;
#relu tends performs better than sigmoid and tanh, however relu returns between
#-1 and 1, so sigmoid will be used in later layers to get output of 0 to 1
#12, 8, and 1 refer to he dimensionality of the output space, ie number of neurons
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#compile model here
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Fit the model
#x/y=input/output; epochs=run training through fixed number of 
#iterations; batch_size = number of instances before synapse weight 
#update
model.fit(X,Y, epochs=1000, batch_size=2)
```

```python
#evaluate scores here, then adjust model. Add more layers,
#change activation function, add more neuron, etc...
scores = model.evaluate(X,Y)
print('\n%s: %.2f%%' % (model.metrics_names[1], scores[1]*100))
''' 
77.6 accuracy was obtained through this model:
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.fit(X,Y, epochs=150, batch_size=10)

95.83 accuracy was obtain through this model
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.fit(X,Y, epochs=1000, batch_size=5)

97.83 accuracy was obtain through this model
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.fit(X,Y, epochs=1000, batch_size=3)

99.48 accuracy was obtain through this model
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.fit(X,Y, epochs=1000, batch_size=2)
'''
```
